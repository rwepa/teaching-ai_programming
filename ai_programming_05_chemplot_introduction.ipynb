## 5.1 chemplot 模組簡介

ChemPlot: a Python Library for Chemical Space Visualization

近代機器學習(Machine Learning, ML)應用對分子和材料科學產生了巨大影響。chemplot 模組特性：(1).提供化學結構的視覺化應用。(2).包含使用結構相似性和定制相似性(structural and tailored similarity)的智慧型演算法。

chemplot 模組五大功能：

1. Similarity Analysis          相似度分析


2. Dimensionality Reduction     維度縮減


3. Visualize the Chemical Space 化學結構視覺化


4. Clustering Analysis          集群分析


5. Hexagonal Bin Plot           六邊箱形圖


6. Kernel Density Estimate Plot 核密度估計圖


參考網址: https://chemplot.readthedocs.io/en/latest/

## 內建9個資料集

## 載入 BBBP 資料集 (blood-brain barrier penetration 血腦屏障穿透)

##### 資料來源： Martins, Ines Filipa, et al. (2012). A Bayesian approach to in silico blood-brain barrier penetration modeling. Journal of chemical information and modeling 52.6, 1686-1697.

```python
from chemplot import load_data
```

```python
df = load_data("BBBP")
```

```python
df # 2039列, 2行
```

```python
from chemplot import Plotter
```

```python
cp = Plotter.from_smiles(df["smiles"], target=df["target"], target_type="C")
```

```python
cp.tsne()
```

## 視覺化-使用 matplotlib 模組

```python
import matplotlib.pyplot as plt
cp.visualize_plot()
```

## 互動式視覺化 - 使用 bokeh 模組

```python
cp.interactive_plot(show_plot=True)
```
